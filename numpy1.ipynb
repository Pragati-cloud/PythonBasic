```python
new1= np.eye(5)
print(new1)
print(new1.size)
print(new1.dtype)
print(new1.sort)
print(new1.itemsize)
print(new1.ndim)
```

```python
a1= np.array([10,20,30,40,50])
a2 = np.array([1,2,3,4,5])

print(a1+a2)
print(a1-a2)
print(a1*a2)
print(a1/a2)
```

```python
a3= np.array([1,2,3,4,5])

print(np.sin(a3))
print(np.log(a3))
print(np.exp(a3))
print(np.sqrt(a3))
```

```python
#array slicing and indexing

arr = np.array([[34,5,6,3,7,3],[45,7,2,9,4,2],[7,3,2,8,6,4]])

print(arr[1:])

```

```python

arr = np.array([[34,5,6,3,7,3],[45,7,2,9,4,2],[7,3,2,8,6,4]])

print(arr[1:][1:])

arr[0][0]=100
print(arr)
```
