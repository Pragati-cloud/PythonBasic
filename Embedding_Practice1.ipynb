```python
import numpy as np

docs = ['hello india','yesterday indian cricket team win world cup','it is very proud moment for india','let cheers for indian team','cheers for india','jemimah , radha , richa , deepti played very well']
```

```python

from tensorflow.keras.preprocessing.text import Tokenizer
tokenizer = Tokenizer(oov_token='<nothing>')
```

```python
tokenizer.fit_on_texts(docs)
```

```python
tokenizer.word_index
```

```python
tokenizer.word_counts
```

```python
tokenizer.document_count
```

```python
sequence = tokenizer.texts_to_sequences(docs)
sequence
```

```python
from keras.utils import pad_sequences
```

```python
sequence = pad_sequences(sequence,padding='post')
sequence
```

```python
from keras import Sequential
from keras.layers import Dense , SimpleRNN, Embedding , Flatten

model = Sequential()
model.add(Embedding(input_dim=25, output_dim=2, input_length=7))
model.build(input_shape=(None,7))
model.summary()
```

```python
model.compile('adam','accuracy')
```

```python
import numpy as np
sequence = np.array(sequence).astype('int32')
print(sequence.shape)   # it MUST print (6,7)
```

```python
pred = model.predict(sequence)
print(pred)
```
