```python
import numpy as np

docs = ['hello india','yesterday indian cricket team win world cup','it is very proud moment for india','let cheers for indian team','cheers for india','jemimah , radha , richa , deepti played very well']
```

```python

from tensorflow.keras.preprocessing.text import Tokenizer
tokenizer = Tokenizer(oov_token='<nothing>')
```

```python
tokenizer.fit_on_texts(docs)
```

```python
tokenizer.word_index
```

```python
tokenizer.word_counts
```

```python
tokenizer.document_count
```

```python
sequence = tokenizer.texts_to_sequences(docs)
sequence
```

```python
from keras.utils import pad_sequences
```

```python
sequence = pad_sequences(sequence,padding='post')
sequence
```
